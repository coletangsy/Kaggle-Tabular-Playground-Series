### Feature Engineering

```python
def datetime_cleaning(df):
    df["date_time"] = pd.to_datetime(df["date_time"])
    df["year"] = df["date_time"].dt.year
    df["month"] = df["date_time"].dt.month
    df["week"] = df["date_time"].dt.week
    df["day"] = df["date_time"].dt.day
    df["dayofweek"] = df["date_time"].dt.dayofweek
    df["time"] = df["date_time"].dt.date - df["date_time"].dt.date.min()
    df["hour"] = df["date_time"].dt.hour
    df["time"] = df["time"].apply(lambda x : x.days)
```

```python
datetime_cleaning(X)
datetime_cleaning(test)
```

```python
def data_encode(df):
    df["dayofweek"] = df["dayofweek"].astype(object) # turn into string to encode
    return pd.get_dummies(df, drop_first = True)
```

```python
X = data_encode(X)
test = data_encode(test)
```

### Apply LazyPredict module to take a look on the regressor

```python
from sklearn.model_selection import train_test_split
from lazypredict.Supervised import LazyRegressor
```

#### Carbon Monoxide

```python
X_train, X_test, y_train, y_test = train_test_split(X, train_co, test_size = 0.2, random_state = 42)
```

```python
reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
models, predictions = reg.fit(X_train, X_test, y_train, y_test)

print(models)
```

From the result of **Lazypredict**, we can observe that **LGBM, HistGradientBoosting and XGB** are the Top 3 well performed regressor for **Carbon Monoxide**.

#### Benzene

```python
X_train, X_test, y_train, y_test = train_test_split(X, train_be, test_size = 0.2, random_state = 42)
```

```python
reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
models, predictions = reg.fit(X_train, X_test, y_train, y_test)

print(models)
```

From the result of **Lazypredict**, we can observe that **MLP, XGB and HistGradientBoosting** are the Top 3 well performed regressor for **Benzene**.

#### Nitrogen Oxides

```python
X_train, X_test, y_train, y_test = train_test_split(X, train_no, test_size = 0.2, random_state = 42)
```

```python
reg = LazyRegressor(verbose=0, ignore_warnings=False, custom_metric=None)
models, predictions = reg.fit(X_train, X_test, y_train, y_test)

print(models)
```

From the result of **Lazypredict**, we can observe that **ExtraTreesRegressor, HistGradientBoosting and XGB** are the Top 3 well performed regressor for **Nitrogen Oxides**.

HistGradientBoosting and XGB regressor both performed well in the Lazypredict, because of my less understanding on HistGradientBoosting now, I would go for the XGB first.

### XGBoost

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_log_error
```

```python
xgb_submission = sample_submission.copy()
```

```python
xgboost = XGBRegressor()

param_dist = {"max_depth": list(range(1,15,2)),
              "min_child_weight" : list(range(1,15,2)),
              "n_estimators": list(range(100,200,20)),
              "learning_rate" : [0.05, 0.1, 0.16]}

grid_search = GridSearchCV(xgboost, 
                           param_grid=param_dist, 
                           cv = 5, 
                           verbose=10, 
                           n_jobs=-1)

grid_search  = grid_search.fit(X_train, train_co)
print(grid_search.best_score_)
print(grid_search.best_params_)
```

```python
xgboost = XGBRegressor(learning_rate = 0.1, 
                       max_depth = 5, 
                       n_estimators  = 180, 
                       min_child_weight = 11,
                       subsample = 0.7,
                       alpha = 0.9) 

xgboost.fit(X_train, train_co) 
xgb_submission['target_carbon_monoxide'] = xgboost.predict(test)
```
